```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
insurance = pd.read_csv("./OneDrive/Desktop/DATA ANALYST/Python/assicurazioni.csv")
```

```python
insurance
```

```python
southwest = insurance[insurance["region"] == "southwest"].shape[0]
southeast = insurance[insurance["region"] == "southeast"].shape[0]
northeast = insurance[insurance["region"] == "northeast"].shape[0]
northwest = insurance[insurance["region"] == "northwest"].shape[0]
```

```python
region = [southwest, southeast, northeast, northwest]
etichette = ["southwest", "southeast", "northeast", "northwest"]
plt.pie(region, labels=etichette)
plt.title("Distribuzione delle persone")
```

```python
fumatori = insurance[insurance["smoker"] == "yes"].shape[0]
non_fumatori = insurance[insurance["smoker"] == "no"].shape[0]
titolo = ["fumatori", "non fumatori"]
dati = [fumatori, non_fumatori]
plt.bar(titolo, dati)
plt.title("quanti fumatori")
plt.ylabel("ammontare")
```

```python
sotto_venti_anni = insurance[(insurance["age"] < 20) & (insurance["sex"] == "male")].shape[0]
sotto_venti_anni_f = insurance[(insurance["age"] < 20) & (insurance["sex"] == "female")].shape[0]
titoli = ["Maschio", "Femmina"]
genere = [sotto_venti_anni, sotto_venti_anni_f]
plt.bar(titoli, genere)
plt.title("Persone sotto i 20 anni assicurate")
```

```python
media_eta = insurance["age"].mean()
media_eta
```

```python
media_importo = insurance["charges"].mean()
media_importo
```

```python
maggiori_30 = insurance[insurance["bmi"] > 30.000].shape[0]
minori_30 = insurance[insurance["bmi"] <= 30.000].shape[0]
bmi = [maggiori_30, minori_30]
legenda = ["maggiori", "minori"]
plt.pie(bmi, labels=legenda)
plt.title("bmi maggiore o minore di 30.000")
```

```python
genitori = insurance [insurance["children"] == 0].shape[0]
non_genitori = insurance [insurance["children"] >= 1].shape[0]
figli = [genitori, non_genitori]
legenda = ["1 figlio o piÃ¹", "0 figli"]
plt.pie(figli, labels=legenda)
plt.title("Genitori o no")
```

```python
importo_medio_southwest = insurance[insurance["region"] == "southwest"]["charges"].mean()
importo_medio_southeast = insurance[insurance["region"] == "southeast"]["charges"].mean()
importo_medio_northwest = insurance[insurance["region"] == "northwest"]["charges"].mean()
importo_medio_northeast = insurance[insurance["region"] == "northeast"]["charges"].mean()
media_regioni = [importo_medio_southwest, importo_medio_southeast, importo_medio_northwest, importo_medio_northeast]
legenda = ["southwest", "southeast", "northwest", "northeast"]
plt.bar(legenda, media_regioni)
plt.title("Importo assicurativo medio per regione")
```
